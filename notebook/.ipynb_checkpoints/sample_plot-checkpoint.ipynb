```python
# example notebook to plot a broad band SED 
# using the data collection used in [2203.11502]
# When using this data collection, please cite the paper:
# Dirson, L. and Horns, D. "Phenomenological..."
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
matplotlib.pyplot.rcParams["figure.figsize"]=(14.5,8)
# i should switch here to units
fpds = 4.0 * np.pi * (2e3 * 3.0857e18)**2 # 4\pi d^2 in cm^2
c_cgs = 2.9979246e+10  # cm/s
cfac = 1e-23 # 1 jansky = 1e-26 W/m^2/Hz = 1e-23 cgs
hev  = 4.1356677e-15
```

```python
# radio data (not corrected for secular changes)
# units: Frequency in GHz, flux(unc) in jansky, epoch in year, reference ADS style
radio_data = pd.read_table("radio_data.csv",delimiter=',',comment='#')
#radio_data.columns
```

```python
# IR data 
# units wlmu in micron, fnu in Jy, fnuunc in Jy
# 
ir_data  = pd.read_table("ir_data.csv", delimiter=',', comment='#')
#ir_data.columns
```

```python
# optical data 
opt_data = pd.read_table("optical_data.csv", delimiter=',', comment='#')
#opt_data.columns
```

```python
# x-ray data
xray_data = pd.read_table("xray_data.csv", delimiter=',', comment='#')
xray_data.columns
```

```python
gamma_data = pd.read_table("gamma_data.csv",delimiter=',', comment='#')
```

```python
# read in model with \alpha=0.51 
datestring = 'Sat_Feb_12_19:06:03_2022_6.dat'
sync = np.genfromtxt('../log/sed_'+datestring)
ic   = np.genfromtxt('../log/sed_IC_'+datestring)
sync_all = np.sum(sync[:,1:4],axis=1)
ic_all = np.sum(ic[:,1:7],axis=1)
data = np.genfromtxt("../notebooks/alldata_v6")
data_kwargs = {'alpha': 1, 'fmt': 'o', 'elinewidth': 0.9, 'ecolor': 'black',
             'mec': 'black', 'mew': 0.6, 'markersize':3, 'color':'b'}



data = data[:-15,:] # cut away the extension stuff for SED
cut = data[:,0]<1e8*2.4e14
sync_data = data[cut,:]
```

```python
plt.xscale("log")
plt.yscale("log")
plt.ylim(5e30,1.5e37)

plt.xlabel(r"$h\nu$ [eV]")
plt.ylabel(r"$\nu L_\nu$ erg/s")
total     = {'color': 'darkgreen', 'linewidth':1.9}
plt.errorbar(hev*radio_data.frequency*1e9,
             radio_data.flux*radio_data.frequency*1e9*fpds*1e-23,
             yerr=radio_data.fluxunc*radio_data.frequency*1e9*fpds*1e-23,fmt='vb')
plt.errorbar(hev*c_cgs*1e4/ir_data.wlmu, ir_data.fnu*c_cgs*1e4/ir_data.wlmu*fpds*cfac, yerr=ir_data.fnuunc*c_cgs/ir_data.wlmu/1e-4*fpds*cfac,fmt='vr')

# this factor is re-correcting the deabsorbed VCV points
reabs = 10**(-0.4 * (opt_data.absorption - opt_data.absorptionvcv))
plt.errorbar(hev*c_cgs*1e8/opt_data.wvlenAng, 
            opt_data.flux * 1e-23*fpds * c_cgs*1e8/opt_data.wvlenAng/ reabs,
            yerr = 0.1 * opt_data.flux*1e-23*fpds/reabs,fmt='^g')
plt.errorbar(xray_data.EkeV*1e3,xray_data.fluxkeVpercgs*fpds*1.602e-9,
            yerr = xray_data.uncertainty*fpds*1.602e-9, fmt='<g')

plt.errorbar(gamma_data.EGeV*1e9,gamma_data.fluxergpercgs*fpds,
            yerr = gamma_data.unc*fpds, fmt='>b')

plt.loglog(sync[:,0],sync_all*sync[:,0]/hev*1e7,**total)
plt.loglog(ic[:,0],ic_all*ic[:,0]/hev*1e7,**total)
```
